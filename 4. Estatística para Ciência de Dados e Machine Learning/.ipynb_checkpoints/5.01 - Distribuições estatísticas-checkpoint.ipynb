## Teorema Central do Limite

Diz que a medida que o tamanho da amostra aumenta, a distribuição amostral da sua média aproxima-se cada vez mais de uma distribuição normal.

```python
alturas = np.random.randint(126, 192, 500)
print(alturas.shape)
srn.histplot(alturas, kde=True, bins=100);
```

```python
n=10000  # tamanho da amostra
medias = [np.mean(np.random.randint(126, 192, 500)) for _ in range(n)]
print(len(medias))
srn.histplot(medias, kde=True, bins=100);
```

### Exemplo - credit data

```python
df_credit = pd.read_csv('Bases de dados/credit_data.csv')
df_credit.dropna(inplace=True)
print(df_credit.shape)
df_credit.head()
```

```python
X = df_credit.iloc[:, 1:4].values
y = df_credit.iloc[:, 4].values
X.shape, y.shape
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
resultados_naive_bayes = []
for i in range(30):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=i)
    
    naive_bayes = GaussianNB()
    naive_bayes.fit(X_train, y_train)
    resultados_naive_bayes.append(accuracy_score(y_test, naive_bayes.predict(X_test)))
resultados_naive_bayes = np.array(resultados_naive_bayes)
```

```python
print(f'Média: {resultados_naive_bayes.mean()}')
print(f'Mediana: {np.median(resultados_naive_bayes)}')
print(f'Moda: {stats.mode(resultados_naive_bayes)}')
print(f'Variância: {np.var(resultados_naive_bayes)}')
print(f'Desvio Padrão: {np.std(resultados_naive_bayes)}')
```

```python
srn.histplot(resultados_naive_bayes, kde=True, bins=10);
```

Quanto mais vc aumentar o tamanho dos resultados, mais ele se aproxima de uma gaussiana, mas agora ele tem um resultado uniforme.

## Variáveis Discretas

## Distribuição de Bernoulli

- Tipo de distribuição binomial 
- Apenas duas respostas

```python
from scipy.stats import bernoulli
```

```python
dados_bernoulli = bernoulli.rvs(size=1000, p=0.8) 
print(dados_bernoulli.shape)
srn.histplot(dados_bernoulli, bins=2);
```

```python
np.unique(dados_bernoulli, return_counts=True)
```

## Distribuição binomial

```python
from scipy.stats import binom
```

```python
dados_binomial = binom.rvs(size=1000, n=10, p=0.5) # gerará n+1 categorias
print(dados_binomial.shape)
srn.histplot(dados_binomial, kde=True, bins=50);
```

```python
np.unique(dados_binomial, return_counts=True)
```
